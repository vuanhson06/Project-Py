```python
import pandas as pd
from sklearn.model_selection import train_test_split

rawdata_path = '../data/raw/sms.csv'  # tạo đường dẫn để biết file data raw của mình ở đâu, rồi mí đọc được
def split_data(rawdata_path):
    df = pd.read_csv(rawdata_path, encoding='latin1')   # đọc đọc đọc

    print(df.head())   # in ra thử 5 dòng để xem đọc đúng chưa...

    X = df['v2']  
    y = df['v1']    # X chứa các đặc trưng (features) là dữ liệu đầu vào dùng để đưa ra dự đoán 
                # y chứa các nhãn or mục tiêu, là đáp án mình muốn nó đoán, là danh sách chứa các nhãn kiểu spam hay ham
                # vì sao phải cắt: ví dụ có 100 tin nhắn, nếu ko cắt mà cho máy học cả thì nó chỉ cần học thuộc lòng full 100 câu, 
                # lúc hỏi lại nó sẽ auto trả lời đúng, thế thì đ có ý nghĩa gì -> nên cần cắt thành 2 tập train và test, vd lấy 80 câu cho học, 
                # học xong lấy 20 câu mới chưa gặp bao giờ ra test -> như vậy mới phản ánh đúng năng lực
    X_train, X_test, y_train, y_test = train_test_split(   # dùng train_test_split để chia dữ liệu
        X,
        y,
        test_size= 0.2,  # này là tỉ lệ cắt, 0.2 tức là 20%, lấy 20% toàn bộ dữ liệu ra làm bài thi, còn lại 80% sẽ tự động đc coi như phần bài tập để học
        random_state= 71,   # cố định sự ngẫu nhiên, ví dụ có 100 test và cần bốc random 20 test ra để làm bài thi, 
                            # nếu ko có random_state thì mỗi lần chạy code nó sẽ bốc ra 20 bài khác nhau, sẽ có lần bốc đc nhiều câu dễ, có lúc khó 
                            # dẫn đến kết quả ko ổn định, vd t chạy thì ra kết quả A nhưng ae cũng chạy cái code này lại ra kqua B, 
                            # nói chung là nó ghi lại quy luật của sự ngẫu nhiên, để ae chạy chung 1 cái code thì ra kết quả như nhau, 
                            # hoặc thử nghiệm các mô hình khác nhau thì nó đc làm chung 1 bộ dữ liệu sẽ dễ so sánh hơn.
        stratify=y          # chia 1 cách công bằng, ví dụ trong 100 tn thì có 90 cái ham và 10 cái spam, 
                            # nếu nó bốc linh tinh 80 test để học đều trúng ham cả thì nó đ biết thế nào là spam -> ko làm đc test, 
                            # khi có stratify=y thì sẽ đảm bảo trong cả 80 câu học và 20 câu thi đều có tỉ lệ spam/ham như nhau.
    )
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = split_data(rawdata_path) #chạy hàm và lưu vào các biến

print(f"number of tests in train: {len(X_train)}")  # này để kiểm tra xem nó chia được chưa hoi
print(f"number of tests in train: {len(X_test)}")
      



```
